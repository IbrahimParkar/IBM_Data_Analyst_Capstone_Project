## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Creating the dataframe 
QUERY = """
SELECT*
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
sns.histplot(df.ConvertedComp)
#alternatively
fig = px.histogram(df, x="ConvertedComp")
fig.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
sns.boxplot(x='Age', data=df);
#Alternatively
box = px.box(df, x='Age')
box.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.scatterplot(x='Age', y='WorkWeekHrs', data = df)
#Alternatively
scatter = px.scatter(df, x='Age', y='WorkWeekHrs')
scatter.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.scatter(x=df.WorkWeekHrs, y=df.CodeRevHrs,alpha=0.2, s=df.Age*10)
#Alternatively
px.scatter(df, x='WorkWeekHrs', y='CodeRevHrs')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query = '''
select * from DatabaseDesireNextYear
'''
df_new=pd.read_sql_query(query, conn)
db_df = df_new.DatabaseDesireNextYear.value_counts().head().reset_index()
db_df = db_df.rename(columns={'index':'Database', 'DatabaseDesireNextYear':'Count'})

#Pie Charts
pie = plt.pie(x=db_df.Count, labels=db_df.Database,autopct='%1.2f%%');
plt.show()
#alternatively
px.pie(db_df, values='Count', names='Database')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
new_df1=df[(df.Age >= 30) & (df.Age <=35)]
new_df1.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().plot.bar(stacked=True)
plt.title('Median hours, by age')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
lin_df = df[(df.Age >= 45) & (df.Age <=60)]
lin_df.groupby('Age')['ConvertedComp'].median().plot.line()
plt.title('Median hours, by age')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df.groupby('MainBranch')['MainBranch'].count().plot(kind='barh', figsize=(10, 6))
plt.title('Horizontal Bar Chart for Main Branch')
```
